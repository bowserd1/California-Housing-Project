# Predicting Median Home Values in California

The booming housing market that many of us are experiencing around the United States is nothing new. Sometimes, looking back at how the housing market has behaved is the best way to understand today's market. This project uses data from California in 1990 and predicts housing prices. While predicting the home prices, the project also looks at what effects a house's price and provides insights not only for these influences in the 1990's but also for the present.

### 2. Feature Engineering

At a quick glance, it is easy to see that the highest priced homes are located in two primary areas: around California's largest cities and along California's coastline. Using the existing coordinate locations, I added features that calculated the distance to the coast for each data point and the distance to the closest major California city with a population greater than 500,000 people. Deciding to limit the cities to ones with populations of at least 500,000 people took some experimenting. At first, I used cities that had over 250,000 people in an attempt to be able to better understand the nuances of the housing market. After looking at the resulting cities on a map, however, the cities that are between 250,000 and 500,000 people are almost all suburbs of the larger cities such as Los Angeles and San Francisco, and thus a part of the same housing market.

## 3. Algorithms and Modelling




#### 1. Preprocessing

To prepare for the modelling stage, the data needs to be manipulated a bit. Since this is tabular data, I plan to try some regression models as well as two decision tree ensemble methods: Random Forest and XG Boost.
In order for the data to be used in the regression models, I scaled all of the numeric data. I also one-hot-encoded the non-numeric data, specifically the city that each housing block was closest to.

I also decided to evaluate the models based on the root mean squared error (RMSE), which will penalize worse predictions.

#### 2. Linear Regression Models
The first model I ran was a basic linear regression model. The training data had an rmse of .695 and it performed worse on the test set. before making more tweaks to this model, I moved on to test other models to see if they had significantly different scores.

#### 3. Random Forest Model
The next model I used was a random forest model. Using cross-validation, the best model used 1000 estimators and returned an improved rmse of .589. Computationally, using 1000 estimators is very expensive and takes an extremely long time for both the training and testing to run.

#### 4. XGBoost Model
The last model I used as an XGBoost. Using crossfold validation, I tuned several hyperparameters to find the best version of the model. Once these had been tuned, the rmse of this model was .57 for both the training set and the test set. This performed the best of any of our models, and once the hyperparameters were tuned, it performed faster than the Random Forest.

## Conclusion

I tried several different models to predict home prices in California. Ultimately, XGBoost provided the best prediction model. The most important feature for prediction was, unsurprisingly, the median income for the housing block. Other features that proved important are the proximity to the Pacific coast and the average number of occupants.
